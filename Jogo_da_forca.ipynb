```python
print("*********BEM VINDO AO JOGO DA FORCA!*********")

palavra_secreta = "banana"
enforcou = False
acertou = False

#Enquanto ainda não enforcou e não acertou
#Não False = True
#Enquanto Não False Enforcou e Não False Acertou
#Enquanto True Enforcou e True False... CONTINUA o processo
while(not enforcou and not acertou):

  chute = input("Qual é a letra? ")
  posicao = 0
  for letra in palavra_secreta:
    if chute == letra:
      print("Encontrei a letra {} na posição {}".format(chute, posicao))
    posicao = posicao + 1

  print("jogando...")
```
